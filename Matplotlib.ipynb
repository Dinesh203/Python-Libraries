# stack plot

```python
days = [1,2,3,4,5]
sleeping = [7,8,6,11,12]
working = [2,3,4,3,2]
playing = [7,8,7,2,3]
eating = [8,5,7,8,13]

plt.plot([], [], color="y", label="sleeping", linewidth=5)
plt.plot([], [], color="c", label="working", linewidth=5)
plt.plot([], [], color="r", label="playing", linewidth=5)
plt.plot([], [], color="b", label="eating", linewidth=5)
plt.stackplot(days, sleeping, working, playing, eating, colors=["y","c","r","b"])
plt.xlabel("X")
plt.ylabel("Y")
plt.title("My scater")
plt.legend()
plt.show()
```

# pie plot

```python
slices = [7,2,3,13]
activites = ['sleeping','working','playing','eating']
cols = ['c','r','b','m']
plt.pie(slices, labels=activites, colors=cols, startangle=90, shadow=True, explode=(0,0.1,0,0), autopct='%1.1f%%')
plt.title("pie plot")
plt.show()
```

# Working with multiple plot

```python
import numpy as np
```

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0,5.0,0.1)
t2 = np.arange(0.0,5.0,0.02)

plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2))
plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2))
plt.show()
```

```python
from ipywidgets import *
```

```python
x = np.linspace(0, 2 * np.pi)

def update(w = 1.0):
    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1)
    ax.plot(x, np.sin(w * x))

    fig.canvas.draw()

interact(update);
```

```python
x = np.linspace(0, 2 * np.pi)
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
line, = ax.plot(x, np.sin(x))

def update(w = 1.0):
    line.set_ydata(np.sin(w * x))
    fig.canvas.draw()

interact(update);
```
